```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('indian_airports.csv')
df.info()
```

```python
plt.bar(df['name'].head(10), df['elevation'].head(10), color='hotpink')  # Plotting only the first 10 for readability
plt.xlabel('Name')
plt.ylabel('Elevation')
plt.title('Elevation of Different Locations')
plt.xticks(rotation=45)
```

```python
# scatter graph of latitude and elevation
c=np.random.randint(1,100, size=len(df))
plt.scatter(df['latitude'],df['elevation'],c=c,cmap='nipy_spectral',alpha=0.8,s=200)
plt.title('Scatter of Airports')
plt.xlabel('latitude')
plt.ylabel('elevation')
```

```python
# Line Graph
plt.plot(df['latitude'], df['elevation'],'o:g',mfc='r')
plt.title('Line Plot of Latitude vs Elevation')
plt.xlabel('Latitude')
plt.ylabel('Elevation')
```

```python
# Histogram
plt.hist(df['elevation'], color='hotpink', edgecolor='r')
plt.title('Histogram of Elevation')
plt.xlabel('Elevation')
plt.ylabel('Frequency')
```

```python
# Box plot
plt.boxplot(df['elevation'])
plt.title('Box Plot of Elevation')
plt.ylabel('Elevation')
```

```python
# Pie Chart
# group by 'location'
location_counts = df['location'].head(50).value_counts()

# Find the index of the maximum value
max_index = location_counts.idxmax()

# Create explode array
myexplode = [0.1 if location == max_index else 0 for location in location_counts.index]

plt.pie(location_counts,explode=myexplode,shadow=True,startangle=90,labels=location_counts.index)
plt.title('Pie Chart of Locations')
plt.xticks(rotation=50)
# plt.legend(labels=location_counts.index,loc='upper right')
```

```python
numeric_columns = ['latitude', 'elevation']
# Matrix of correlations
correlation_matrix = df[numeric_columns].corr()
plt.imshow(correlation_matrix,cmap='autumn')
plt.title('Correlation Heatmap')
```

```python
# Area plot
df[ 'elevation'].plot.area()
plt.title('Area Plot of Elevation')
plt.xlabel('Index')
plt.ylabel('Value')
```

```python
# stem plot
plt.stem(df.index, df['latitude'])
plt.title('Stem Plot of Latitude')
plt.xlabel('Index')
plt.ylabel('Latitude')
```

```python
# step plot 
plt.step(df.index, df['elevation'], where='mid')
plt.title('Step Plot of Elevation')
plt.xlabel('Index')
plt.ylabel('Elevation')
plt.grid(True)
```

```python
# 3D plot of Latitude and Elevation and Last Updated(timestamp)
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111, projection='3d')
df['last_updated'] = pd.to_datetime(df['last_updated'])
# Convert last_updated to a numerical format (e.g., timestamp)
df['last_updated_timestamp'] = df['last_updated'].map(pd.Timestamp.timestamp)

ax.scatter(df['latitude'], df['elevation'], df['last_updated_timestamp'], c='y', marker='o', label='Latitude',s=450)
ax.scatter(df['latitude'], df['elevation'], df['last_updated_timestamp'], c='r', marker='o', label='Elevation',alpha=0.4,s=200)
ax.scatter(df['latitude'], df['elevation'], df['last_updated_timestamp'], c='c', marker='o', label='last_updated_timestamp',alpha=0.6,s=30)

# all 3 in one
# scatter = ax.scatter(df['latitude'], df['elevation'], df['last_updated_numeric'], c='b', marker='*')

ax.set_xlabel('Latitude')
ax.set_ylabel('Elevation')
ax.set_zlabel('timestamp')
ax.set_title('3D Scatter Plot with Latitude and Elevation and Last Updated')
plt.legend(loc='upper left')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(1, 20, size=15)
y = np.random.randint(1, 20, size=15)

# Create a 3x3 grid of subplots
plt.figure(figsize=(12, 12))

# Plot 1: Line plot
plt.subplot(3, 3, 1)
plt.plot(x, y, '*:g')
plt.title('Line Plot')

# Plot 2: Scatter plot
plt.subplot(3, 3, 2)
plt.scatter(x, y, color='r')
plt.title('Scatter Plot')

# Plot 3: Histogram
plt.subplot(3, 3, 3)
plt.hist(x, color='b')
plt.title('Histogram')

# Plot 4: Bar plot
plt.subplot(3, 3, 4)
plt.bar(x, y, color='yellow')
plt.title('Bar Plot')

# Plot 5: Another type of plot (Example: Hexbin)
plt.subplot(3, 3, 5)
plt.hexbin(x, y, gridsize=10, cmap='viridis')
plt.title('Hexbin Plot')

# Plot 6: Another type of plot (Example: Boxplot)
plt.subplot(3, 3, 6)
plt.boxplot(x)
plt.title('Box Plot')

# Plot 7: Another type of plot (Example: Violin plot)
plt.subplot(3, 3, 7)
plt.violinplot(y)
plt.title('Violin Plot')

# Plot 8: Another type of plot (Example: Pie chart)
plt.subplot(3, 3, 8)
plt.pie(x)
plt.title('Pie Chart')

# Plot 9: Another type of plot (Example: Area plot)
plt.subplot(3, 3, 9)
plt.fill_between(x, y, color='orange', alpha=0.3)
plt.title('Area Plot')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
